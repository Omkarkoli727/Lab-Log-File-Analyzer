```python
from collections import Counter, defaultdict

input_file = input("Enter input file name: ")

# create output file name automatically
output_file = input_file.split(".")[0] + "_out.txt"

student_time = defaultdict(int)
activity_count = Counter()
lab_time = defaultdict(int)

with open(input_file, "r") as f:
    for line in f:
        line = line.strip()

        if line == "":
            continue

        data = line.split(",")

        if len(data) != 4 or data[3].lower() == "duration":
            continue

        date = data[0].strip()
        name = data[1].strip()
        activity = data[2].strip()

        try:
            duration = int(data[3].strip())
        except:
            continue

        student_time[name] += duration
        activity_count[activity] += 1

        if activity.lower() == "lab":
            lab_time[date] += duration

most_activity = activity_count.most_common(1)[0]

with open(output_file, "w") as out:
    out.write("===== STUDENT ACTIVITY SUMMARY =====\n\n")

    out.write("Total Time by Students:\n")
    for s, t in student_time.items():
        out.write(f"{s} : {t} minutes\n")

    out.write("\nMost Common Activity:\n")
    out.write(f"{most_activity[0]} ({most_activity[1]} times)\n")

    out.write("\nLab Time Per Day:\n")
    for d, t in lab_time.items():
        out.write(f"{d} : {t} minutes\n")

    out.write("\n==================================\n")

print(f"Output file created: {output_file}")
```
